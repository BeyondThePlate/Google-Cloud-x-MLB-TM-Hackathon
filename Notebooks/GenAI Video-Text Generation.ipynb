```python
import google.generativeai as genai
from IPython.display import Markdown
import pandas as pd
import os
```

```python
from google.colab import userdata


genai.configure(api_key='API Key')
```

```python
video_file_name = "/content/HomeRunTest.mp4"

print(f"Uploading file...")
video_file = genai.upload_file(path=video_file_name)
print(f"Completed upload: {video_file.uri}")
```

```python
import time

# Check whether the file is ready to be used.
while video_file.state.name == "PROCESSING":
    print('.', end='')
    time.sleep(10)
    video_file = genai.get_file(video_file.name)

if video_file.state.name == "FAILED":
  raise ValueError(video_file.state.name)
```

```python
# Create the prompt.
prompt = "Summarize this video. Then create a quiz with answer key based on the information in the video."

# Choose a Gemini model.
model = genai.GenerativeModel(model_name="gemini-1.5-pro-latest")

# Make the LLM request.
print("Making LLM inference request...")
response = model.generate_content([video_file, prompt],
                                  request_options={"timeout": 600})

# Print the response, rendering any Markdown
Markdown(response.text)
```

```python
for video in videos:
  #os.system(f"!wget {video}")

  video = video.replace("https://sporty-clips.mlb.com/","")

  print(video)

  video_file_name = f"/content/{video}"


  video_file = genai.upload_file(path=video_file_name)

  time.sleep(10)

  # Create the prompt.
  prompt = f"""Summarize the video in 4 sentences.
  Provide only the description itself, without any introductions, labels, or additional commentary."""


 # Choose a Gemini model.
  model = genai.GenerativeModel(model_name="gemini-1.5-pro-latest")

 # Make the LLM request.
  print("Making LLM inference request...")
  response = model.generate_content([video_file, prompt],
                                  request_options={"timeout": 600})

 # Print the response, rendering any Markdown
  Markdown(response.text)

  gen_data.loc[len(gen_data)] = response.text
  #gen_data = pd.concat([response.text,gen_data], ignore_index=True)

```
